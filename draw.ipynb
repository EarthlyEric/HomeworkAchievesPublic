```python
# type: ignore
import pandas 
from matplotlib import pyplot
from sklearn.linear_model import LinearRegression

dataset = pandas.read_csv('production_data.csv')
P_quantity = dataset.iloc[:, 1].values.reshape(-1, 1)
P_expense = dataset.iloc[:, 2].values.reshape(-1, 1)

model = LinearRegression()
model.fit(P_quantity, P_expense)
predicted_P_expense = model.predict(P_quantity)

pyplot.scatter(P_quantity, P_expense, label='Data Points')
pyplot.plot(P_quantity, predicted_P_expense, color='red',label='Linear Regression Line')
pyplot.xlabel('P_quantity')
pyplot.ylabel('P_expense')
pyplot.legend()
pyplot.show()
```
